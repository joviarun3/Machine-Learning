```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset, drop_first=True)
#dataset = dataset.astype(int)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
#Here we are not gonna split test & train data. Becasue it is done by cross validation itself in GridSearch CV method.
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid = {'kernel':['rbf','poly','sigmoid','linear'],'C':[1,0.0001,0.01,0.1,2.0],'gamma':['auto','scale']}
grid = GridSearchCV(SVR(), param_grid, refit = True, verbose = 3,n_jobs=-1)
#In the above line we are loading different combination of given hyper tuning parameters
grid.fit(independent, dependent)
```

```python
re=grid.cv_results_#this "cv_results" function will store all the values from GridSearchCV into variable "re"
print("TheR_score valueforbestparameter{}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re) #"DataFrame.from_dict" will tablate the all the given hypertuning values.
table
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children:"))
sex_male_input=int(input("Sex Male 0 or 1:"))
smoker_yes_input=int(input("Smoker Yes 0 or 1:"))
```

```python
Future_prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
Future_prediction
```
