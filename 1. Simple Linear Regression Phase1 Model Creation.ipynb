```python
#pandas is a library that contains all functions related to the database. we are importing all functions from pandas library as 'pd' for easy use (to avoid lenghty name)
import pandas as pd
```

```python
#we read all data from the dataset file (csv file recievd from customer) using read_csv function from pandas library and assigned to variabke called dataset,
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#just callling datset variabke to view & verify what is assigned in it.
dataset
```

```python
#Assigning the 'Years of experience' values from dataset to the variable called independent (i.e) Input X
#Additional note: multiple columns can be stored in a variable with ' as separator.for example, independent = [["colum1 name","column2 name",column3 name".....]]
independent=dataset[["YearsExperience"]]
independent
```

```python
#Assigning the 'Salary' values from dataset to the variable called dependent (i.e) Output y
dependent=dataset[["Salary"]]
dependent
```

```python
#we import 'train_test_split' function from 'model_selection' class from 'sklearn' library
#sklearn is commonly used for machine learning.
#this function will split the input and output data for 'training' and 'testing' in the required size ratio e.g 0.30
#It means 30% data will be used only for testing the created model and 70% of data will be used for training the model (Training is also called learning)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
#we import 'LinearRegression' function from 'linear_model' class from 'sklearn' library
from sklearn.linear_model import LinearRegression
#assigning all functions from LinearRegression to variable called regressor. So it becomes easy to call LinearRegression function as regressor now.
regressor=LinearRegression()
#Now regressor will have all formulae, functions etc..
#'fit' is used to find (w) weight & (b) bais values here. We are just substituing the training input and training output to find w and b.
regressor.fit(X_train,y_train)
```

```python
#just to view what is the value of w (.coef_ is used for this)
weight=regressor.coef_
weight
```

```python
#just to view what is the value of b (.intercept_ used for this)
bais=regressor.intercept_
bais
```

```python
# we have created the model with training data. now we are going to test the created model with test data (30% of data that we split above)
# we use predict function from the regressor to predict the output of test input data (note: test data was not part of train data)
#we store the predicted o/p data in y_pred varaible
y_pred=regressor.predict(X_test)
```

```python
#evaluation metrics. We are using 4th method(R2 Squared) to compare the predicted output of test data with actual output of test data
#we import 'r2_score' function from 'metrics' class from 'sklearn' library
#then the result of evaluation is assigned to a library called r_score. This method will give result from 0 to 1. 
#1 or closer to 1 means model has learned very good.
#0 or closed to 0 means it is not a good model
#Calling the r_score variable to view the evalation value.
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
#Now our model is good and best..so need to save it.
#pickle is one of the libraries that we use here for saving the model.
#importing pickle library and assignig a file name with .sav extention and then calling dump function from pickle
#best practise is saving the model name with key words "finalized" and "type of algorithm"
#dump function in pickle needs "which data" needs to written in "which file". 'regressor' has all the model related data.
import pickle
filename="finalized_model_linear.sav"
pickle.dump(regressor,open(filename, 'wb'))#wb is write binary
```

```python
#we are going to load the saved model and check if it works
#using load function from pickle library and reading the file the assigning to variable called loaded model.
#No we have everything loaded in the model.Using predict function to get the predicted output. It just needs the input. (i.e) X
loaded_model=pickle.load(open("finalized_model_linear.sav", 'rb'))#rb is read binary
result=loaded_model.predict([[13]])
result
```
